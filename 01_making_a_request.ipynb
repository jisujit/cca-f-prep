```python
import os
from dotenv import load_dotenv

load_dotenv()
print(os.getenv("ANTHROPIC_API_KEY")[:8])  # Only prints first 8 chars for safety
```

```python
import anthropic

client = anthropic.Anthropic()

message = client.messages.create(
    model="claude-opus-4-5",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hello, Claude! from Sujit!!"}
    ]
)
print(message.content)
```

```python
print(message.content[1])
```

```python
print(message[1])
```

```python
print(message.content[0].text)
```

```python
from anthropic.types import TextBlock
for item in message.content: 
      if isinstance(item,TextBlock):
          print(item.text)
```

```python
for item in message.content:
    if item.type == "text":
        print(item.text)
        
```
